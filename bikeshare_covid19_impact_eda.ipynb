```python
#reading datasets from 2017 2018 2019 2020 2021 and 2022 
import pandas as pd
import glob

# Get a list of all CSV files in the directory
file_list = glob.glob('F:\DataSciCAB\DataSci\Bike\Bike-DS\cov19\*.csv')

# Create an empty list to hold the DataFrames
df_list = []

# Loop through the CSV files and read them into DataFrames
for file in file_list:
    # Read the CSV file into a DataFrame and rename the columns
    df = pd.read_csv(file)
    df = df.rename(columns={'ride_id': 'Ride ID', 'rideable_type': 'Rideable Type',
                            'started_at': 'Start date', 'ended_at': 'End date',
                            'start_station_name': 'Start station', 'start_station_id': 'Start station number',
                            'end_station_name': 'End station', 'end_station_id': 'End station number',
                            'start_lat': 'Start station latitude', 'start_lng': 'Start station longitude',
                            'end_lat': 'End station latitude', 'end_lng': 'End station longitude',
                            'Duration': 'Duration (seconds)', 'Bike number': 'Bike Number',
                            'member_casual': 'Member type'})
   
    df_list.append(df)

# Concatenate the DataFrames
df = pd.concat(df_list, ignore_index=True)

```

```python
#we have 18441 duplicated rows so to get unique rows we can drop them
df = df.drop_duplicates(keep='first')
```

```python
#selecting columns that we need 
bikes = df[['Start date', 'End date', 'Member type']] 
```

```python
#to see the type of columns 
bikes.info()
```

```python
# convert start_time and end_time to datetime objects and member type to category 
import warnings
warnings.filterwarnings('ignore')

bikes['Start date'] = pd.to_datetime(bikes['Start date'])
bikes['End date'] = pd.to_datetime(bikes['End date'])
bikes['Member type'] = bikes['Member type'].astype('category')
```

```python
#make year month day time and hour features
bikes['Month']=bikes['Start date'].dt.month
bikes['Year']=bikes['Start date'].dt.year
bikes['day'] = bikes['Start date'].dt.day
bikes['day_name'] = bikes['Start date'].dt.day_name()
bikes['time'] = bikes['Start date'].dt.time
bikes['hour'] = bikes['Start date'].dt.hour

#change the day names to categorical type
bikes['day_name'] = bikes['day_name'].astype('category')
```

```python
#to get seasons from months 

#a function to map month numbers to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

# Create a new categorical column called "seasons"
bikes['seasons'] = bikes['Month'].apply(get_season).astype('category')
```

```python
#check for missing values 
bikes.isna().sum()
```

```python
#to change the values from Member to registered member and from casual to casual member
bikes['Member type'] = bikes['Member type'].replace({'Member': 'registered member', 'Casual': 'casual member', 'member': 'registered member', 'casual': 'casual member'})
```

```python
# calculate duration in seconds
bikes['Duration'] = (bikes['End date'] - bikes['Start date']).dt.total_seconds()
```

```python
#rename columns in bikes ds
for col in bikes.columns:
    bikes.rename(columns={col:col.lower().replace(" ","_")},inplace=True)

bikes.head(0)
```

```python
#to check if we have negative value for duration or not
bikes[bikes['duration'] < 0].shape
```

Sunday and Monday, suggesting that the bikes were being used for daily commutes or regular work/school activities.
This trend may have changed during and after the pandemic, as people's work and travel habits were affected.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define the time categories
time_bins = [0, 6, 12, 17, 20, 24]
time_labels = ['Night', 'Morning', 'Noon', 'Afternoon', 'Evening']

# Create a new column 'time_category' based on the 'hr' column
bikes['time_category'] = pd.cut(bikes['hour'], bins=time_bins, labels=time_labels, include_lowest=True)

# Group the data by time_category and calculate the count of bike rentals for each category
hourly_counts = bikes.groupby(['time_category', 'year']).size().reset_index(name='count')

# Create a pivot table to reformat the data for the heatmap
hourly_pivot = hourly_counts.pivot(index='time_category', columns='year', values='count')

# Define a custom function to format the annotations
def format_label(value):
    if value > 1000000:
        return ">1M"
    elif value > 1000:
        return str(int(value/1000)) + "K"
    else:
        return str(int(value))

# Create a new figure and set the size of the plot
plt.figure(figsize=(10, 6))

# Create a heatmap of the hourly trends of bike rentals by time category
sns.heatmap(hourly_pivot, cmap='Reds', fmt='', annot_kws={'size': 12}, cbar=False, xticklabels=True, yticklabels=True, linewidths=1, linecolor='white')

# Loop through the annotations and set the text to the abbreviated labels
for i in range(len(hourly_pivot)):
    for j in range(len(hourly_pivot.columns)):
        text = hourly_pivot.iloc[i, j]
        plt.text(j + 0.5, i + 0.5, format_label(text), ha='center', va='center', size=12)

plt.xlabel('Period')
plt.ylabel('Time Category')
plt.title('Hourly Trends of Bike Rentals by Time Category (2017-2022)')

# Show the plot
plt.show()
```

```python
# Convert the year column to numeric format
bikes['year'] = pd.to_numeric(bikes['year'])

# Get the weekday and weekend durations for each year
years = bikes['year'].unique()

weekday_durations = []
weekend_durations = []

for year in years:
    year_data = bikes[bikes['year'] == year]
    weekday_durations.append(year_data[year_data['day_name'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]['duration'])
    weekend_durations.append(year_data[year_data['day_name'].isin(['Saturday', 'Sunday'])]['duration'])

# Set the style and color of the box and whiskers
boxprops = dict(linestyle='-', linewidth=2, color='darkblue')
whiskerprops = dict(linestyle='--', linewidth=1.5, color='gray')
capprops = dict(linestyle='-', linewidth=2, color='black')
medianprops = dict(linestyle='-', linewidth=2.5, color='red')

# Create a box plot of the weekday and weekend durations for each year
fig, ax = plt.subplots(figsize=(12, 8))

bp = ax.boxplot(weekday_durations, positions=[x - 0.2 for x in years], widths=0.3,
                boxprops=boxprops, whiskerprops=whiskerprops, capprops=capprops, medianprops=medianprops,
                showfliers=False, patch_artist=True)

for patch in bp['boxes']:
    patch.set_facecolor('lightblue')

bp = ax.boxplot(weekend_durations, positions=[x + 0.2 for x in years], widths=0.3,
                boxprops=boxprops, whiskerprops=whiskerprops, capprops=capprops, medianprops=medianprops,
                showfliers=False, patch_artist=True)

for patch in bp['boxes']:
    patch.set_facecolor('pink')

# Add labels and title
plt.xlabel('Year', fontsize=14)
plt.ylabel('Duration (sec)', fontsize=14)
plt.title('Distribution of bikesharing ride durations by day type for each year', fontsize=16)

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add x-axis tick labels
plt.xticks(years, years)

# Add legend
weekday_patch = plt.Rectangle((0, 0), 1, 1, fc="lightblue")
weekend_patch = plt.Rectangle((0, 0), 1, 1, fc="pink")
plt.legend([weekday_patch, weekend_patch], ['Weekdays', 'Weekends'], loc='upper left')

plt.show()
```

Observation: duration of bike rides were longer in the weekends compared to weekdays before during and after pandemic.

To check for numbers of registered and casual users 

```python
# Calculate the total count of registered and casual users for each period
period_counts = bikes.groupby(['year', 'member_type'])['member_type'].count().unstack()
period_counts = period_counts.fillna(0)

# Calculate the percentage of registered and casual users for each period
period_counts['total'] = period_counts.sum(axis=1)
period_counts['Registered Users'] = (period_counts['registered member'] / period_counts['total']) * 100
period_counts['Casual Users'] = (period_counts['casual member'] / period_counts['total']) * 100

# Create a single plot with a pie chart for each period
fig, axes = plt.subplots(nrows=1, ncols=6, figsize=(12, 4))

for i, period in enumerate(period_counts.index.tolist()):
    labels = ['Registered Users', 'Casual Users']
    sizes = [period_counts.loc[period]['Registered Users'], period_counts.loc[period]['Casual Users']]
    colors = ['lightblue', 'lightgreen']
    ax = axes[i]
    ax.pie(sizes, colors=colors, autopct='%1.1f%%', startangle=90)
    ax.set_title(period)

fig.suptitle('Percentage of Registered and Casual Users for Each Period')
ax.legend(labels,  loc='lower right', bbox_to_anchor=(0.5, -0.3), ncol=1, frameon=False)

plt.show()
```

Observation: the propotion of casual members increased during and after pandemic.
